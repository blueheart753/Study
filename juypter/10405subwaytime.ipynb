```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('seoul.csv','r',encoding = 'cp949')
data = csv.reader(f)

next(data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')  # ['1907','10','01']
        if 2000 <= int(date[0]):
            if date[1] == '12' and date[2] == '25' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
X = list(range(0,40,2))
ticklabel = [2000 + i for i in X]
plt.xticks(X, ticklabel)
plt.rc('font', family = 'Malgun Gothic')
plt.title('2000년 부터 2년 단위의 크리스마스 기온 변화 그래프')
plt.rcParams['axes.unicode_minus'] = False
plt.plot(high, 'hotpink', label = 'high')
plt.plot(low,'skyblue', label = 'low')
plt.legend()
plt.show()
```
