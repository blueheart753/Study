```python
f = open('seoul.csv','r',encoding = 'cp949')
data = csv.reader(f)
max = -999
max_date = ''
next(data)
for row in data:
    if row[-1] == '':
        row[-1] == -9999
    if max < float(row[-1]):
        max = float(row[-1])         
        max_date = row[0]
print('서울의 가장 높았던 기온은',max,'도 입니다.')
print('서울의 가장 높았던 날은'+max_date+'입니다.')
f.close()
```

```python
f = open('seoul.csv','r',encoding = 'cp949')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
print(result)
f.close()
```

```python
f = open('seoul.csv','r',encoding = 'cp949')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '10' and row[0].split('-')[2]=='25':
            result.append(float(row[-1]))
f.close()

plt.figure(figsize=(10,2))
plt.plot(result,'hotpink')
```
