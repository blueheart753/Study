**변수란?**

컴퓨터에서 변수(Variable)은 데이터가 저장된 컴퓨터 메모리의 위치


'a=5' 는 같다란 의미가 아니라 오른쪽의 데이터를 왼쪽의 변수에 저장하는 것

**변수의 데이터형** 


1.   정수(Integer)
2.   실수(Float)
3.   불런(Boolean)
4.   문자열(String)



[정수] 자연수, -0, 23,0,-17

```python
num = 1
num2 = 2
num3 = 3

print(num,num2,num3)
```

[실수] 17.3, -22.56, 3.0 3.14등 소수점을 가진 숫자

```python
num = 17.3
num2 = -22
num3 = 56

print(num,num2,num3)
```

[불런] 참과 거짓, (True, False)

```python
a = True
b = False

print(a,b)
```

[문자열] "안녕하세요", "사과", "a", "school", "010-111-1111"

```python
apple = "apple"

print(apple)
```

 **변수명 만들기**
 
*   영어 대소문자
*   밑줄(_)
*   숫자

```python
num = '010-2931-2157'

print(num)
```
