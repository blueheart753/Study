### 3) 리스트에 반복문 for 사용하기

```python
animal = ["팽귄","토끼","여우","원숭이","사슴"]

for a in animal :
  print("내가 좋아하는 동물 : %s"%a)
```

다섯 과목의 성적의 합계와 평균 구하기

```python
scores = [ 80, 92, 99, 75, 87]
sum = 0
avg = 0

for i in scores:
  sum += i

avg = sum/len(scores)
print("합계 : %d"%sum,"평균 : %.2f"%avg)
```

### 4)리스트에 while문 사용하기

```python
scores = [ 80, 92, 99, 75, 87]
sum = 0
i = 0
while i <len(scores):
  sum += scores[i]
  i = i+1
avg = sum/len(scores)
print("점수 :%d 평균 :2.f%"%(sum,avg))
```

###7) 연습문제

키보드로 성적을 입력 받아 리스트에 저장한 다음 평균을 구하는 프로그램을 작성하시오. 단, -1을 입력했을 때 성적 입력을 종료함.

실행결과

성적을 입력하세요(종료 시 -1 입력) : 55

성적을 입력하세요(종료 시 -1 입력) : 33

성적을 입력하세요(종료 시 -1 입력) : 33

성적을 입력하세요(종료 시 -1 입력) : 55

성적을 입력하세요(종료 시 -1 입력) : -1

입력된 성적 :  [55, 33, 33, 55]

평균 : 44.00

```python
score = []
while True:
  scores = int(input("성적을 입력하세요(종료시 -1 입력) :"))
  if scores == -1:
    break
  score.append(scores)
sum = 0
for score in scores :
  sum = sum+score
avg = sum/len(score)
print("입력된 성적 : ",score)
print("평균 : ",avg)
```
