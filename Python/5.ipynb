#명함 만들기

필요한 라이브러리 3종 세트

```python
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
```



---



명함 사이즈에 맞게 캔버스 그리고 text 넣기

```python
x = 1039
y = 697
```

```python
canvas = Image.new ("RGBA",(x,y),"yellow")
canvas.save("test.png")
```

```python
namefont = ImageFont.truetype("C/window/font/gulim.ttc",70)
smallfont= ImageFont.truetype("C/window//font/gulim.ttc",70)
```

```python
ImageDraw.Draw(canvas).text(xy=(0,0),text="Hello world!", font = namefont, fill = "black")
ImageDraw.Draw(canvas).text(xy=(0,0),text="12345", font = smallfont, fill = "black")
```

```python
canvas.save("test2.png")
```
